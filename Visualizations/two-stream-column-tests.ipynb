```python
# Python code to get the Cumulative sum of a list 
def Cumulative(lists): 
    cu_list = [] 
    length = len(lists) 
    cu_list = [sum(lists[0:x:1]) for x in range(0, length+1)] 
    return cu_list[1:]
 
def print_list(values):
    print('[' + ', '.join(format(value, '.3f') for value in values) + ']')

def interpolate(inp, fi):
    i, f = int(fi // 1), fi % 1  # Split floating-point index into whole & fractional parts.
    j = i+1 if f > 0 else i  # Avoid index error.
    return (1-f) * inp[i] + f * inp[j]

real_values = [1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.48797509e-05, 7.82163028e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.27920416e-05, 7.91229077e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.09613417e-05, 8.16930012e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.90971384e-05, 8.13579662e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.72956541e-05, 8.19851572e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.55474639e-05, 8.25180986e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.38138320e-05, 8.26636810e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.02202881e-04, 8.38904921e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.10626016e-04, 8.42313514e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.19167332e-04, 8.54131609e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.27857349e-04, 8.69001681e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.36820391e-04, 8.96304159e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.45812851e-04, 8.99246016e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.54961846e-04, 9.14899501e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.64270966e-04, 9.30912029e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.73743235e-04, 9.47226891e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.83507423e-04, 9.76418815e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.93305144e-04, 9.79772072e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.03247850e-04, 9.94270570e-06,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.13354705e-04, 1.01068547e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.23777767e-04, 1.04230625e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.34239072e-04, 1.04613054e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.44882776e-04, 1.06437032e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.55710433e-04, 1.08276571e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.66725580e-04, 1.10151471e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.78084863e-04, 1.13592832e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.89483954e-04, 1.13990907e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.01080454e-04, 1.15965002e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.12876829e-04, 1.17963756e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.25040094e-04, 1.21632649e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.37246817e-04, 1.22067227e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.49664522e-04, 1.24177046e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.62296993e-04, 1.26324716e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.75148489e-04, 1.28514955e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.88401179e-04, 1.32526904e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.01700756e-04, 1.32995773e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.15230252e-04, 1.35294960e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.28994616e-04, 1.37643637e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.43188278e-04, 1.41936616e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.57432560e-04, 1.42442826e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.71924093e-04, 1.44915328e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.86666160e-04, 1.47420671e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.01663406e-04, 1.49972458e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.17129008e-04, 1.54656017e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.32650523e-04, 1.55215148e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.48441100e-04, 1.57905776e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.64504916e-04, 1.60638159e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.80847958e-04, 1.63430417e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.97702463e-04, 1.68545048e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.14617453e-04, 1.69149898e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.31825787e-04, 1.72083349e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.49332657e-04, 1.75068694e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.67386877e-04, 1.80542200e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.85505530e-04, 1.81186529e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.03938866e-04, 1.84333359e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.22692452e-04, 1.87535861e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.41772019e-04, 1.90795672e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.61450868e-04, 1.96788487e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.81201684e-04, 1.97508168e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.01296819e-04, 2.00951341e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.21742131e-04, 2.04453125e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.42829497e-04, 2.10873659e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.63994181e-04, 2.11646840e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.85531641e-04, 2.15374601e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.07454314e-04, 2.19226729e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.29776885e-04, 2.23225710e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.52815528e-04, 2.30386432e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.75958409e-04, 2.31428808e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.99523101e-04, 2.35646917e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.02351653e-03, 2.39934310e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.04828025e-03, 2.47637177e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.07317927e-03, 2.48990204e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.09862777e-03, 2.54485049e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.12470928e-03, 2.60815081e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.15143735e-03, 2.67280678e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.17918490e-03, 2.77475462e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.20723258e-03, 2.80476796e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.23594917e-03, 2.87165934e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.26533696e-03, 2.93877923e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.29578914e-03, 3.04521826e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.32653293e-03, 3.07437840e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.35796206e-03, 3.14291334e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.39009339e-03, 3.21313267e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.42297035e-03, 3.28769619e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.45704713e-03, 3.40767833e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.49148632e-03, 3.44391909e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.52672315e-03, 3.52368257e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.56276086e-03, 3.60377058e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.60000864e-03, 3.72477827e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.63752217e-03, 3.75135350e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.67575530e-03, 3.82331240e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.71471865e-03, 3.89633494e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.75442163e-03, 3.97029802e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.79537194e-03, 4.09503105e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.83659802e-03, 4.12260779e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.87860306e-03, 4.20050420e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.92140294e-03, 4.27998805e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.96553826e-03, 4.41353189e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.00995630e-03, 4.44180382e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.05517128e-03, 4.52149796e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.10118027e-03, 4.60089963e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.14799527e-03, 4.68149951e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.19620653e-03, 4.82112611e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.24467148e-03, 4.84649496e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.29398416e-03, 4.93126816e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.34416625e-03, 5.01820954e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.39584912e-03, 5.16828616e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.44781188e-03, 5.19627628e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.50068486e-03, 5.28729777e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.55448095e-03, 5.37960988e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.60921231e-03, 5.47313513e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.66557090e-03, 5.63585903e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.72222102e-03, 5.66501199e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.77985566e-03, 5.76346433e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.83848946e-03, 5.86338036e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.89886467e-03, 6.03752037e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.95955043e-03, 6.06857584e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.02128908e-03, 6.17386544e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.08409599e-03, 6.28069055e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.14798761e-03, 6.38916245e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.21377365e-03, 6.57860422e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.27990715e-03, 6.61334955e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.34719965e-03, 6.72925072e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.41566985e-03, 6.84701997e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.48617963e-03, 7.05097784e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.55705924e-03, 7.08796036e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.62917115e-03, 7.21119103e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.70253565e-03, 7.33645023e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.77717269e-03, 7.46370440e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.85402288e-03, 7.68501848e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.93127066e-03, 7.72477826e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.00986023e-03, 7.85895656e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.08982025e-03, 7.99600284e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.17215964e-03, 8.23393873e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.25492281e-03, 8.27631641e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.33912179e-03, 8.41989860e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.42478023e-03, 8.56584410e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.51193706e-03, 8.71568301e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.60170290e-03, 8.97658390e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.69194671e-03, 9.02438071e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.78377397e-03, 9.18272630e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.87720546e-03, 9.34314853e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.97341977e-03, 9.62143106e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.07013955e-03, 9.67197792e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.16854255e-03, 9.84030063e-05,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.26866170e-03, 1.00119151e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.37051942e-03, 1.01857715e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.47541357e-03, 1.04894151e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.58086974e-03, 1.05456173e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.68816645e-03, 1.07296708e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.79733412e-03, 1.09167668e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.90974808e-03, 1.12413964e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.02274605e-03, 1.12997963e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.13771876e-03, 1.14972711e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.25470758e-03, 1.16988825e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.37374668e-03, 1.19039096e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.49634015e-03, 1.22593469e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.61958023e-03, 1.23240084e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.74497220e-03, 1.25391964e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.87254832e-03, 1.27576127e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.00391889e-03, 1.31370566e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.13596587e-03, 1.32046983e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.27030772e-03, 1.34341845e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.40699164e-03, 1.36683922e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.54606032e-03, 1.39068680e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.68927150e-03, 1.43211178e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.83322248e-03, 1.43950979e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.97967763e-03, 1.46455153e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.12866413e-03, 1.48986503e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.28207740e-03, 1.53413271e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.43630499e-03, 1.54227583e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.59323409e-03, 1.56929104e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.75290680e-03, 1.59672712e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.91536799e-03, 1.62461192e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.08266391e-03, 1.67295914e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.25083119e-03, 1.68167285e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.42191959e-03, 1.71088393e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.59597365e-03, 1.74054067e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.77519069e-03, 1.79217035e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.95536160e-03, 1.80170913e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.01386898e-02, 1.83328207e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.03252261e-02, 1.86536303e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.05150178e-02, 1.89791708e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.07104627e-02, 1.95444881e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.09069319e-02, 1.96469205e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.11068217e-02, 1.99889840e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.13102017e-02, 2.03379999e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.15196614e-02, 2.09459606e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.17302367e-02, 2.10575398e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.19444991e-02, 2.14262380e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.21625038e-02, 2.18004648e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.23843089e-02, 2.21805120e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.26127065e-02, 2.28397589e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.28422926e-02, 2.29586137e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.30758609e-02, 2.33568240e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.33135149e-02, 2.37654034e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.35582622e-02, 2.44747351e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.38043077e-02, 2.46045406e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.40546387e-02, 2.50331043e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.43093283e-02, 2.54689599e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.45684503e-02, 2.59122025e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.48352604e-02, 2.66810107e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.51034718e-02, 2.68211357e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.53763726e-02, 2.72900863e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.56540608e-02, 2.77688167e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.59400232e-02, 2.85962398e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.62274858e-02, 2.87462603e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.65199534e-02, 2.92467625e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.68175062e-02, 2.97552756e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.71202162e-02, 3.02710039e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.74319053e-02, 3.11689102e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.77452626e-02, 3.13357259e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.80641150e-02, 3.18852359e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.83885565e-02, 3.24441520e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.87226632e-02, 3.34106736e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.90585360e-02, 3.35872770e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.94002578e-02, 3.41721825e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.97479352e-02, 3.47677357e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.01016527e-02, 3.53717551e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.04659298e-02, 3.64277039e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.08321521e-02, 3.66222308e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.12047813e-02, 3.72629246e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.15839102e-02, 3.79128910e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.19743328e-02, 3.90422563e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.23667883e-02, 3.92455556e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.27660711e-02, 3.99282749e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.31722708e-02, 4.06199695e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.35855818e-02, 4.13310976e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.40112450e-02, 4.25663227e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.44392014e-02, 4.27956382e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.48746423e-02, 4.35440928e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.53177081e-02, 4.43065757e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.57739585e-02, 4.56250403e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.62325990e-02, 4.58640593e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.66992244e-02, 4.66625322e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.71739870e-02, 4.74762665e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.76570724e-02, 4.83085347e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.81599299e-02, 5.02857533e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.86731990e-02, 5.13269130e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.92030541e-02, 5.29855049e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.97497365e-02, 5.46682387e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.03198302e-02, 5.70093711e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.09007902e-02, 5.80960049e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.14993056e-02, 5.98515373e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.21156754e-02, 6.16369750e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.27501079e-02, 6.34432518e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.34097718e-02, 6.59663872e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.40809785e-02, 6.71206706e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.47709180e-02, 6.89939511e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.54798358e-02, 7.08917805e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.62154750e-02, 7.35639260e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.69605781e-02, 7.45103047e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.77196629e-02, 7.59084799e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.84929283e-02, 7.73265405e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.92805560e-02, 7.87627682e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.00908958e-02, 8.10339870e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.09079035e-02, 8.17007627e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.17398745e-02, 8.31970990e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.25870180e-02, 8.47143506e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.34496382e-02, 8.62620189e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.43597124e-02, 9.10074273e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.53043031e-02, 9.44590639e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.62923413e-02, 9.88038217e-04,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.73238151e-02, 1.03147385e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.84080266e-02, 1.08421144e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.95257455e-02, 1.11771889e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.06862506e-02, 1.16050514e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.18899060e-02, 1.20365542e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.31369213e-02, 1.24701526e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.44374699e-02, 1.30054858e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.57705471e-02, 1.33307720e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.71464651e-02, 1.37591801e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.85652286e-02, 1.41876356e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.00377169e-02, 1.47248826e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.15411783e-02, 1.50346136e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.31355772e-02, 1.59439895e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.49623918e-02, 1.82681459e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.70189014e-02, 2.05650963e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.93202754e-02, 2.30137402e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.18451723e-02, 2.52489687e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.46086444e-02, 2.76347211e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.76136846e-02, 3.00504021e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.08762672e-02, 3.26258262e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.43743891e-02, 3.49812190e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.81244166e-02, 3.75002749e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.21295293e-02, 4.00511269e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.63928767e-02, 4.26334742e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.00931531e-01, 4.53865436e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.05720686e-01, 4.78915510e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.10777469e-01, 5.05678293e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.16104908e-01, 5.32743908e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.21717590e-01, 5.61268165e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.27444229e-01, 5.72663948e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.33298063e-01, 5.85383366e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.39280034e-01, 5.98197090e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.45391451e-01, 6.11141739e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.51649896e-01, 6.25844506e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.58023745e-01, 6.37384901e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.64530651e-01, 6.50690514e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.71173082e-01, 6.64243148e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.77970588e-01, 6.79750604e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.84889169e-01, 6.91858055e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.91947286e-01, 7.05811694e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.99146582e-01, 7.19929680e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.06487463e-01, 7.34088115e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.13990889e-01, 7.50342566e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.21619651e-01, 7.62876149e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.29393947e-01, 7.77429669e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.37333633e-01, 7.93968526e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.45470089e-01, 8.13645656e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.53763405e-01, 8.29331569e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.62235785e-01, 8.47238017e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.70888811e-01, 8.65302639e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.79724160e-01, 8.83534825e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.88765966e-01, 9.04180645e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.97973353e-01, 9.20738684e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.07370562e-01, 9.39720933e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.16959164e-01, 9.58860134e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.26765029e-01, 9.80586567e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.36754381e-01, 9.98935130e-03,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.46952714e-01, 1.01983339e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.57361395e-01, 1.04086807e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.67981551e-01, 1.06201555e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.78841484e-01, 1.08599338e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.89904778e-01, 1.10632937e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.01223745e-01, 1.13189669e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.12800024e-01, 1.15762795e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.24663818e-01, 1.18637932e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.36760930e-01, 1.20971128e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.49121060e-01, 1.23601300e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.61746865e-01, 1.26258043e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.74643609e-01, 1.28967443e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.87843972e-01, 1.32003635e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.01289267e-01, 1.34452940e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.15015320e-01, 1.37260535e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.29084566e-01, 1.40692459e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.43532274e-01, 1.44477078e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.58268256e-01, 1.47359825e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.73326123e-01, 1.50578671e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.88707931e-01, 1.53818080e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.04418647e-01, 1.57107156e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.20544943e-01, 1.61262967e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.37016980e-01, 1.64720362e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.53872262e-01, 1.68552826e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.71112043e-01, 1.72397808e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.88776785e-01, 1.76647419e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.06792504e-01, 1.80157196e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.25204520e-01, 1.84120152e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.44015069e-01, 1.88105491e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.63225714e-01, 1.92106456e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.82880823e-01, 1.96551086e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.02891876e-01, 2.00110535e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.23257911e-01, 2.03660342e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.43974338e-01, 2.07164272e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.65088537e-01, 2.11141995e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.86511317e-01, 2.14227793e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.08289645e-01, 2.17783287e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.30471588e-01, 2.21819430e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.53080685e-01, 2.26090969e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.76168326e-01, 2.30876408e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.99636437e-01, 2.34681111e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.02353639e+00, 2.38999490e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.04787723e+00, 2.43408481e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.07272076e+00, 2.48435300e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.09796267e+00, 2.52419074e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.12365833e+00, 2.56956560e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.14980972e+00, 2.61513921e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.17641991e+00, 2.66101909e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.20355223e+00, 2.71323206e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.23109984e+00, 2.75476116e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.25912399e+00, 2.80241476e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.28762682e+00, 2.85028342e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.31667379e+00, 2.90469650e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.34616877e+00, 2.94949834e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.37623643e+00, 3.00676545e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.40687937e+00, 3.06429450e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.43809961e+00, 3.12202415e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.46996884e+00, 3.18692301e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.50237067e+00, 3.24018229e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.53538217e+00, 3.30115043e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.56900485e+00, 3.36226826e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.60331810e+00, 3.43132459e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.63817372e+00, 3.48556170e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.67364969e+00, 3.54759722e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.70974828e+00, 3.60985953e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.74647452e+00, 3.67262338e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.78392165e+00, 3.74471359e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.82193104e+00, 3.80093874e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.86058771e+00, 3.86566709e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.89990271e+00, 3.93149977e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.94005880e+00, 4.01560906e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.98089576e+00, 4.08369584e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.02250494e+00, 4.16091817e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.06500951e+00, 4.25045689e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.10897878e+00, 4.39692698e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.15463348e+00, 4.56547009e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.20193977e+00, 4.73062943e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.25102539e+00, 4.90856197e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.30191196e+00, 5.08865666e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.35471972e+00, 5.28077601e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.40927583e+00, 5.45561099e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.46571137e+00, 5.64355355e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.52405654e+00, 5.83451704e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.58434888e+00, 6.02923442e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.64673083e+00, 6.23819479e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.71100903e+00, 6.42782003e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.77741417e+00, 6.64051452e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.84602268e+00, 6.86085060e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.91698684e+00, 7.09641562e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.98967462e+00, 7.26877832e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.06398926e+00, 7.43146423e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.13994785e+00, 7.59585832e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.21756925e+00, 7.76214061e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.29699989e+00, 7.94306371e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.37799034e+00, 8.09904529e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.46068528e+00, 8.26949370e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.54510900e+00, 8.44237175e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.63141921e+00, 8.63102143e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.71934988e+00, 8.79306682e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.80905947e+00, 8.97095883e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.90057901e+00, 9.15195433e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.99392300e+00, 9.33439943e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.08934211e+00, 9.54191019e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.18663960e+00, 9.72974917e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.28602772e+00, 9.93881236e-02,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.38770421e+00, 1.01676492e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.49187119e+00, 1.04166974e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.59821013e+00, 1.06338940e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.70690760e+00, 1.08697469e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.81798709e+00, 1.11079492e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.93146354e+00, 1.13476447e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.04753990e+00, 1.16076367e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.16586711e+00, 1.18327208e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.28665123e+00, 1.20784123e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.40990933e+00, 1.23258092e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.53586800e+00, 1.25958678e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.66415760e+00, 1.28289593e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.79501050e+00, 1.30852905e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.92851374e+00, 1.33503238e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.06482823e+00, 1.36314488e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.20447106e+00, 1.39642833e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.34710730e+00, 1.42636242e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.49298177e+00, 1.45874467e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.64212461e+00, 1.49142844e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.79480105e+00, 1.52676437e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.95056841e+00, 1.55767362e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.10968814e+00, 1.59119731e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.27218940e+00, 1.62501252e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.43810260e+00, 1.65913206e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.60772576e+00, 1.69623161e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.78056964e+00, 1.72843880e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.95691907e+00, 1.76349429e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.13680440e+00, 1.79885330e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.32055490e+00, 1.83750500e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.50773168e+00, 1.87176776e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.69925202e+00, 1.91520339e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.89546531e+00, 1.96213292e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.09640373e+00, 2.00938417e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.30241659e+00, 2.06012862e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.51291790e+00, 2.10501309e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.72825170e+00, 2.15333802e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.94844561e+00, 2.20193909e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.01738726e+01, 2.25426968e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.04039070e+01, 2.30034386e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.06389064e+01, 2.34999435e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.08789059e+01, 2.39999522e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.11239431e+01, 2.45037151e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.13744257e+01, 2.50482600e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.16296590e+01, 2.55233328e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.18900370e+01, 2.60378034e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.21562289e+01, 2.66191876e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.24315599e+01, 2.75331027e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.27165034e+01, 2.84943466e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.30102200e+01, 2.93716562e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.33130734e+01, 3.02853437e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.36250339e+01, 3.11960470e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.39465104e+01, 3.21476556e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.42766123e+01, 3.30101900e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.46157364e+01, 3.39124117e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.49638569e+01, 3.48120501e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.53209430e+01, 3.57086079e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.56874545e+01, 3.66511524e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.60624206e+01, 3.74966025e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.64462867e+01, 3.83866078e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.68390244e+01, 3.92737706e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.72411290e+01, 4.02104617e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.76523731e+01, 4.11244102e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.80773474e+01, 4.24974288e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.85161047e+01, 4.38757378e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.89686982e+01, 4.52593479e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.94357359e+01, 4.67037635e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.99161522e+01, 4.80416337e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.04105408e+01, 4.94388625e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.09189531e+01, 5.08412329e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.14420455e+01, 5.23092390e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.19786597e+01, 5.36614182e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.25294330e+01, 5.50773240e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.30944056e+01, 5.64972681e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.36736331e+01, 5.79227490e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.42678149e+01, 5.94181735e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.48756837e+01, 6.07868807e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.54979140e+01, 6.22230305e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.61350454e+01, 6.37131460e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.67863054e+01, 6.51259997e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.74503425e+01, 6.64037086e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.81279088e+01, 6.77566307e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.88190608e+01, 6.91151999e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.95238617e+01, 7.04800922e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.02431494e+01, 7.19287611e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.09754284e+01, 7.32279042e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.17215324e+01, 7.46103976e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.24815233e+01, 7.59990937e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.32563011e+01, 7.74777776e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.40442570e+01, 7.87955905e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.48463015e+01, 8.02044523e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.56624707e+01, 8.16169151e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.64928455e+01, 8.30374854e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.73383870e+01, 8.45541452e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.81979027e+01, 8.59515722e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.90763508e+01, 8.78448099e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.99745159e+01, 8.98165137e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.08934403e+01, 9.18924319e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.18312340e+01, 9.37793732e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.27889324e+01, 9.57698436e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.37665984e+01, 9.77665977e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.47642811e+01, 9.97682657e-01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.57831407e+01, 1.01885966e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.68210950e+01, 1.03795425e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.78792701e+01, 1.05817516e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.89577364e+01, 1.07846631e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.00577173e+01, 1.09998087e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.11769888e+01, 1.11927151e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.23167481e+01, 1.13975932e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.34770568e+01, 1.16030862e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.46581212e+01, 1.18106447e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.58649168e+01, 1.20679557e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.70950375e+01, 1.23012068e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.83498255e+01, 1.25478800e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.96293673e+01, 1.27954185e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.09351587e+01, 1.30579138e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.22645739e+01, 1.32941518e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.36190903e+01, 1.35451641e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.49988092e+01, 1.37971893e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.64038393e+01, 1.40503008e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.78358145e+01, 1.43197518e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.92918601e+01, 1.45604559e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.07735885e+01, 1.48172843e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.22811083e+01, 1.50751977e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.38161518e+01, 1.53504353e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.53756596e+01, 1.55950783e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.69613469e+01, 1.58568725e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.85769271e+01, 1.61558029e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.02261917e+01, 1.64926453e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.19111327e+01, 1.68494098e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.36284207e+01, 1.71728806e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.53799226e+01, 1.75150191e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.71658302e+01, 1.78590756e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.89881891e+01, 1.82235893e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.08434054e+01, 1.85521623e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.27334741e+01, 1.89006873e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.46586059e+01, 1.92513178e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.66189431e+01, 1.96033719e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.86167961e+01, 1.99785309e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.00648116e+02, 2.03131987e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.02715178e+02, 2.06706220e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.04818132e+02, 2.10295365e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.06959432e+02, 2.14130003e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.09137291e+02, 2.17785955e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.11361331e+02, 2.22403906e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.13631963e+02, 2.27063239e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.15949429e+02, 2.31746660e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.18316465e+02, 2.36703543e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.20728304e+02, 2.41183937e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.23187688e+02, 2.45938380e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.25694812e+02, 2.50712419e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.28252648e+02, 2.55783536e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.30856038e+02, 2.60339078e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.33507912e+02, 2.65187357e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.36208500e+02, 2.70058810e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.38958037e+02, 2.74953677e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.41759813e+02, 2.80177580e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.44608128e+02, 2.84831543e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.47506172e+02, 2.89804348e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.50455851e+02, 2.94967913e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.53471909e+02, 3.01605807e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.56548442e+02, 3.07653313e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.59689014e+02, 3.14057247e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.62893876e+02, 3.20486117e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.66163447e+02, 3.26957122e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.69501529e+02, 3.33808178e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.72901505e+02, 3.39997650e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.76367029e+02, 3.46552379e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.79898611e+02, 3.53158178e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.83500352e+02, 3.60174096e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.87165051e+02, 3.66469962e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.90896723e+02, 3.73167155e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.94695789e+02, 3.79906667e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.98562561e+02, 3.86677178e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.02501758e+02, 3.93919670e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.06506436e+02, 4.00467800e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.10596196e+02, 4.08975967e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.14771597e+02, 4.17540148e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.19037709e+02, 4.26611213e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.23385873e+02, 4.34816332e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.27821002e+02, 4.43512947e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.32343595e+02, 4.52259290e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.36954084e+02, 4.61048879e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.41657912e+02, 4.70382807e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.46445648e+02, 4.78773600e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.51322716e+02, 4.87706840e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.56289447e+02, 4.96673130e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.61351819e+02, 5.06237182e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.66499644e+02, 5.14782484e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.71738672e+02, 5.23902817e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.77069363e+02, 5.33069109e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.82492245e+02, 5.42288153e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.87987454e+02, 5.49520864e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.93544984e+02, 5.55753014e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.99168964e+02, 5.62398079e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.04857893e+02, 5.68892816e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.10616335e+02, 5.75844229e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.16430326e+02, 5.81399161e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.22304356e+02, 5.87402951e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.28236869e+02, 5.93251303e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.34226166e+02, 5.98929702e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.40277476e+02, 6.05130973e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.46375508e+02, 6.09803176e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.52525329e+02, 6.14982129e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.58725199e+02, 6.19986979e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.64980798e+02, 6.25559916e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.71275740e+02, 6.29494183e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.77615628e+02, 6.33988812e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.83983849e+02, 6.36822103e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.90246256e+02, 6.26240702e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.96412438e+02, 6.16618246e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.02468204e+02, 6.05576528e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.08423243e+02, 5.95503949e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.14279268e+02, 5.85602449e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.20046416e+02, 5.76714808e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.25709607e+02, 5.66319173e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.31279145e+02, 5.56953731e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.36756619e+02, 5.47747422e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.42143955e+02, 5.38733598e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.47452157e+02, 5.30820223e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.52664829e+02, 5.21267219e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.57792975e+02, 5.12814610e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.62838453e+02, 5.04547732e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.67812934e+02, 4.97448091e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.72699024e+02, 4.88609013e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.77507897e+02, 4.80887305e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.82396866e+02, 4.88896906e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.87381472e+02, 4.98460600e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.92472560e+02, 5.09108830e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.97650096e+02, 5.17753586e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.02925273e+02, 5.27517680e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.08298425e+02, 5.37315248e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.13781802e+02, 5.48337681e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.19353613e+02, 5.57181053e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.25025863e+02, 5.67225006e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.30798871e+02, 5.77300833e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.36673601e+02, 5.87472968e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.42663147e+02, 5.98954589e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.48743567e+02, 6.08042017e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.54927946e+02, 6.18437891e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.61217228e+02, 6.28928229e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.67625151e+02, 6.40792348e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.74126233e+02, 6.50108174e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.80736599e+02, 6.61036559e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.87475369e+02, 6.73877053e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 5.94343709e+02, 6.86833976e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.01356570e+02, 7.01286075e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.08486925e+02, 7.13035532e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.15749668e+02, 7.26274281e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.23145659e+02, 7.39599101e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.30691200e+02, 7.54554142e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.38357049e+02, 7.66584821e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.46159318e+02, 7.80226974e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.54098853e+02, 7.93953423e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.62177141e+02, 8.07828815e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.70411288e+02, 8.23414700e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.78770031e+02, 8.35874296e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.87270798e+02, 8.50076708e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 6.95914542e+02, 8.64374401e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.04702584e+02, 8.78804213e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.13653973e+02, 8.95138949e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.22734421e+02, 9.08044797e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.31982127e+02, 9.24770612e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.41406630e+02, 9.42450302e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.51028246e+02, 9.62161600e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.60810716e+02, 9.78247000e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.70774135e+02, 9.96341803e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.80920220e+02, 1.01460856e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 7.91250180e+02, 1.03299600e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.01786163e+02, 1.05359825e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.12488965e+02, 1.07028024e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.23380217e+02, 1.08912525e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.34461557e+02, 1.10813390e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.45756528e+02, 1.12949716e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.57223317e+02, 1.14667889e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.68884995e+02, 1.16616779e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.80743300e+02, 1.18583052e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 8.92799882e+02, 1.20565816e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.05080670e+02, 1.22807880e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.17539892e+02, 1.24592223e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.30202515e+02, 1.26626225e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.43095922e+02, 1.28934078e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.56259081e+02, 1.31631586e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.69643966e+02, 1.33848848e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.83277151e+02, 1.36331852e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 9.97161071e+02, 1.38839202e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.01129810e+03, 1.41370300e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.02571828e+03, 1.44201772e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.04036976e+03, 1.46514804e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.05528132e+03, 1.49115598e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.07045553e+03, 1.51742094e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.08592431e+03, 1.54687843e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.10163236e+03, 1.57080518e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.11761039e+03, 1.59780256e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.13386139e+03, 1.62509986e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.15038759e+03, 1.65262051e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.16722395e+03, 1.68363587e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.18431046e+03, 1.70865121e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.20168025e+03, 1.73697879e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.21933672e+03, 1.76564632e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.23731843e+03, 1.79817123e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.25560652e+03, 1.82880875e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.27424093e+03, 1.86344174e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.29322498e+03, 1.89840516e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.31258033e+03, 1.93553463e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.33228180e+03, 1.97014684e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.35234402e+03, 2.00622185e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.37277080e+03, 2.04267844e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.39356972e+03, 2.07989136e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.41473724e+03, 2.11675272e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.43628053e+03, 2.15432853e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.45820402e+03, 2.19234963e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.48051206e+03, 2.23080364e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.50321249e+03, 2.27004342e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.52630192e+03, 2.30894268e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.54978926e+03, 2.34873424e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.57367910e+03, 2.38898394e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.59798049e+03, 2.43013870e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.62268953e+03, 2.47090374e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.64781513e+03, 2.51255974e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.67336241e+03, 2.55472855e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.69937326e+03, 2.60108519e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.72589134e+03, 2.65180809e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.75291298e+03, 2.70216402e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.78044886e+03, 2.75358789e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.80850565e+03, 2.80567882e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.83709436e+03, 2.85887072e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.86621150e+03, 2.91171467e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.89586919e+03, 2.96576833e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.92607403e+03, 3.02048452e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.95683316e+03, 3.07591257e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 1.98815904e+03, 3.13258769e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.02004859e+03, 3.18895521e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.05251503e+03, 3.24664456e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.08556610e+03, 3.30510620e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.11920988e+03, 3.36437847e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.15346055e+03, 3.42506723e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.18831416e+03, 3.48536112e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.22378525e+03, 3.54710850e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.25988316e+03, 3.60979159e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.29662387e+03, 3.67407016e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.33400564e+03, 3.73817771e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.37145142e+03, 3.74457722e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.40766812e+03, 3.62167025e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.44264289e+03, 3.49747723e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.47636934e+03, 3.37264458e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.50881974e+03, 3.24504066e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.53998668e+03, 3.11669404e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.56985549e+03, 2.98688115e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.59841951e+03, 2.85640192e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.62564757e+03, 2.72280587e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.65153123e+03, 2.58836551e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.67605408e+03, 2.45228515e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.69919911e+03, 2.31450265e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.72095736e+03, 2.17582509e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.74129435e+03, 2.03369921e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.76019967e+03, 1.89053254e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.77765441e+03, 1.74547350e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.79363860e+03, 1.59841877e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.80814036e+03, 1.45017648e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.82112117e+03, 1.29808094e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.83256755e+03, 1.14463774e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.84245676e+03, 9.88921025e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.85077288e+03, 8.31612069e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.85798521e+03, 7.21233202e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.86533219e+03, 7.34697767e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.87281708e+03, 7.48488945e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.88044378e+03, 7.62670567e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.88822673e+03, 7.78294694e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.89614941e+03, 7.92267610e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.90422721e+03, 8.07780608e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.91246458e+03, 8.23736667e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.92087749e+03, 8.41290675e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.92944792e+03, 8.57043270e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.93819237e+03, 8.74445147e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.94711574e+03, 8.92336743e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.95622396e+03, 9.10821969e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.96553538e+03, 9.31142475e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.97503053e+03, 9.49514998e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.98472871e+03, 9.69817498e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 2.99463641e+03, 9.90770574e+00,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.00476068e+03, 1.01242629e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.01512330e+03, 1.03626192e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.02570371e+03, 1.05804191e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.03652438e+03, 1.08206607e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.04759400e+03, 1.10696197e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.05893767e+03, 1.13436710e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.07053200e+03, 1.15943304e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.08239885e+03, 1.18668514e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.09454928e+03, 1.21504283e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.10699496e+03, 1.24456800e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.11976591e+03, 1.27709510e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.13284053e+03, 1.30746190e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.14625077e+03, 1.34102404e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.16001236e+03, 1.37615955e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.17416326e+03, 1.41509020e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.18868275e+03, 1.45194858e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.20361022e+03, 1.49274680e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.21896688e+03, 1.53566591e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.23477644e+03, 1.58095663e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.25108737e+03, 1.63109301e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.26788330e+03, 1.67959269e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.28521787e+03, 1.73345734e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.30312626e+03, 1.79083888e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.32164699e+03, 1.85207307e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.34085062e+03, 1.92036278e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.36073424e+03, 1.98836189e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.38137991e+03, 2.06456745e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.40285145e+03, 2.14715332e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.42525435e+03, 2.24029067e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.44861448e+03, 2.33601273e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.47305661e+03, 2.44421288e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.49869472e+03, 2.56381109e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.52567430e+03, 2.69795825e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.55421325e+03, 2.85389450e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.58445669e+03, 3.02434431e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.61672996e+03, 3.22732731e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.65140956e+03, 3.46795935e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.68901412e+03, 3.76045632e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.73034700e+03, 4.13328829e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.77643389e+03, 4.60868884e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.82923597e+03, 5.28020804e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.89245450e+03, 6.32185254e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 3.97573595e+03, 8.32814540e+01,
1.50000000e+03, 9.90000000e-01, 9.90000000e-01, 4.17845163e+03, 2.02715677e+02]

FLUXES = {}
THERMAL_FLUXES = {}
NUS = np.logspace(13, 15, 100)
W0s = [0.0, 0.15, 0.30, 0.45, 0.60, 0.75, 0.90]
G0s = [-0.99, 0.0, 0.99]

for W0_VAL in W0s:
    for G0_VAL in G0s:
        W0_G0 = str(W0_VAL) + str(G0_VAL)
        FLUXES[W0_G0] = []
        THERMAL_FLUXES[W0_G0] = []

        for NU in NUS:
            NEW_NLAYERS = [len(real_values) / 5]

            final_vals1 = []
            final_vals2 = []
            for q in range(len(NEW_NLAYERS)):
                NEW_NLAYER = int(NEW_NLAYERS[q])

                W0 = []
                G0 = []
                TEMPS = []
                TAULS = []
                TAUCS = []

                for i in range(0, len(real_values), 5):
                    TEMPS.append(real_values[i])

                    W0.append(W0_VAL)
                    G0.append(G0_VAL)

                    #W0.append(real_values[i+1])
                    #G0.append(real_values[i+2])

                    TAUCS.append(real_values[i+3])
                    TAULS.append(real_values[i+4])


                #old_NLAYER = len(real_values) / 4
                #TEMPS = [interpolate(TEMPS, i*(len(TEMPS)-1) / (NEW_NLAYER-1)) for i in range(NEW_NLAYER)]
                #W0    = [interpolate(W0, i*(len(W0)-1) / (NEW_NLAYER-1)) for i in range(NEW_NLAYER)]
                #G0    = [interpolate(G0, i*(len(G0)-1) / (NEW_NLAYER-1)) for i in range(NEW_NLAYER)]
                #TAULS = np.asarray([interpolate(TAULS, i*(len(TAULS)-1) / (NEW_NLAYER-1)) for i in range(NEW_NLAYER)])*(old_NLAYER/NEW_NLAYER)
                
                mu_0 = 1.0
                mu = 1.0

                bolz_constant = 1.380649e-23
                h_constant    = 6.62607015e-34

                CLIGHT = 3e8
                STELLAR_TEMP = 6213
                incident_frac = 1
                PI = np.pi
                R_STAR = 1.029e09
                ORB_SEP = 8.901e+09

                DIRECT_HEMISPHERIC= [None] * NEW_NLAYER
                DIRECT_QUADRATURE= [None] * NEW_NLAYER

                y1= [None] * NEW_NLAYER
                y2= [None] * NEW_NLAYER
                y3= [None] * NEW_NLAYER
                y4= [None] * NEW_NLAYER

                LAMBDAS= [None] * NEW_NLAYER
                GAMMA= [None] * NEW_NLAYER
                temp_e_val= [None] * NEW_NLAYER
                e1= [None] * NEW_NLAYER
                e2= [None] * NEW_NLAYER
                e3= [None] * NEW_NLAYER
                e4= [None] * NEW_NLAYER

                A = [None] * 2 * NEW_NLAYER
                B = [None] * 2 * NEW_NLAYER
                D = [None] * 2 * NEW_NLAYER
                E = [None] * 2 * NEW_NLAYER

                AS= [None] * 2 * NEW_NLAYER
                DS= [None] * 2 * NEW_NLAYER
                X= [None] * 2 * NEW_NLAYER
                Y= [None] * 2 * NEW_NLAYER

                temp_gamma_val= [None] * NEW_NLAYER

                B0= [None] * NEW_NLAYER
                B1= [None] * NEW_NLAYER
                TEMP= [None] * NEW_NLAYER
                DTAUS= [None] * NEW_NLAYER

                CP= [None] * NEW_NLAYER
                CPB= [None] * NEW_NLAYER
                CM= [None] * NEW_NLAYER
                CMB= [None] * NEW_NLAYER

                SOURCE_G= [None] * NEW_NLAYER
                SOURCE_H= [None] * NEW_NLAYER
                SOURCE_J= [None] * NEW_NLAYER
                SOURCE_K= [None] * NEW_NLAYER
                ALPHA_1= [None] * NEW_NLAYER
                ALPHA_2= [None] * NEW_NLAYER
                SIGMA_1= [None] * NEW_NLAYER
                SIGMA_2= [None] * NEW_NLAYER
                SOURCE_Y1= [None] * NEW_NLAYER
                SOURCE_Y2= [None] * NEW_NLAYER
                source_temp= [None] * NEW_NLAYER

                INTENSITY_DOWN= [None] * NEW_NLAYER
                INTENSITY_UP= [None] * NEW_NLAYER
                TMI  = [None] * NEW_NLAYER
                FLUX = [None] * NEW_NLAYER

                HEMISPHERIC_TWO_STREAM= [None] * NEW_NLAYER
                HEMISPHERIC_SOURCE_FNC= [None] * NEW_NLAYER

                QUADRATURE_TWO_STREAM= [None] * NEW_NLAYER
                QUADRATURE_SOURCE_FNC= [None] * NEW_NLAYER

                temp_val_1 = (2.0 * h_constant * (NU * NU * NU)) / (CLIGHT * CLIGHT);
                temp_val_2 = np.exp(h_constant * NU / (bolz_constant * STELLAR_TEMP)) - 1.0;
                STELLAR_BB = temp_val_1 * (1.0 / temp_val_2);

                FLUX_SURFACE_HEMISPHERIC = 0;
                FLUX_SURFACE_QUADRATURE = incident_frac * PI * STELLAR_BB * pow((R_STAR / ORB_SEP), 2.0);

                temp_val_1 = (2.0 * h_constant * (NU * NU * NU)) / (CLIGHT * CLIGHT);
                temp_val_2 = np.exp(h_constant * NU / (bolz_constant * TEMPS[NEW_NLAYER-1])) - 1.0;
                BB_TOP_OF_ATM = temp_val_1 * (1.0 / temp_val_2);

                temp_val_1 = (2.0 * h_constant * (NU * NU * NU)) / (CLIGHT * CLIGHT);
                temp_val_2 = np.exp(h_constant * NU / (bolz_constant * TEMPS[NEW_NLAYER-1])) - 1.0;
                BB_BOTTOM_OF_ATM = temp_val_1 * (1.0 / temp_val_2);

                mu_1 = 0.5;
                EMIS = 1.0;
                RSFX = 1.0;

                SFCS_HEMISPHERIC = EMIS * PI * BB_BOTTOM_OF_ATM;
                
            
                for J in range(NEW_NLAYER):
                    DIRECT_QUADRATURE[J]  = mu_0 * PI * FLUX_SURFACE_QUADRATURE * np.exp(-1.0 * (TAUCS[J] + TAULS[J]) / mu_0);
                    DIRECT_HEMISPHERIC[J] = 0.0;

                for J in range(NEW_NLAYER):
                    y1[J]    =  (2.0 - (W0[J] * (1.0 + G0[J])));
                    y2[J]    =  (W0[J] * (1.0 - G0[J]));

                    LAMBDAS[J]    =  (y1[J] ** 2.0 - y2[J] ** 2.0) ** 0.5
                    GAMMA[J]  =  y2[J] / (y1[J] + LAMBDAS[J]);
                    temp_e_val[J]   =  np.exp(-LAMBDAS[J] * TAULS[J]);

                    e1[J]   =  1.0 + GAMMA[J] * temp_e_val[J];                         
                    e2[J]   =  1.0 - GAMMA[J] * temp_e_val[J];                    
                    e3[J]   =  GAMMA[J] + temp_e_val[J];                      
                    e4[J]   =  GAMMA[J] - temp_e_val[J];

                J = 0;
                for L in range(1, 2*NEW_NLAYER-1, 2):
                    # HERE ARE THE EVEN MATRIX ELEMENTS
                    A[L]   =  e2[J+1] * e1[J]   - e3[J] * e4[J+1];
                    B[L]   =  e2[J+1] * e2[J]   - e4[J+1] * e4[J];
                    D[L]   =  e1[J+1] * e4[J+1] - e3[J+1] * e2[J+1];

                    # HERE ARE THE ODD MATRIX ELEMENTS EXCEPT FOR THE TOP. 
                    A[L+1] =  e2[J]   * e3[J]   - e1[J]   * e4[J];
                    B[L+1] =  e1[J+1] * e1[J]   - e3[J+1] * e3[J]; 
                    D[L+1] =  e3[J]   * e4[J+1] - e1[J]   * e2[J+1];
                    J = J + 1;


                # HERE ARE THE TOP AND BOTTOM BOUNDARY CONDITIONS AS WELL AS THE
                # BEGINNING OF THE TRIDIAGONAL SOLUTION DDINITIONS. I ASSUME
                # NO DIFFUSE RADIATION IS INCIDENT AT UPPER BOUNDARY.
                A[0] = 0.0;
                B[0] = e1[0];
                D[0] = -e2[0];

                A[2*NEW_NLAYER-1] = e1[NEW_NLAYER-1] - RSFX * e3[NEW_NLAYER-1];
                B[2*NEW_NLAYER-1] = e2[NEW_NLAYER-1] - RSFX * e4[NEW_NLAYER-1];
                D[2*NEW_NLAYER-1] = 0.0;

                # This is the part of the code that solves for the blackbody stuff
                for J in range(NEW_NLAYER):
                    if(0 >= J-1):
                        KINDEX = 0;
                    else:
                        KINDEX = J-1;

                    temp_val_1 = (2.0 * h_constant * (NU * NU * NU)) / (CLIGHT * CLIGHT);
                    temp_val_2 = np.exp(h_constant * NU / (bolz_constant * TEMPS[J])) - 1.0;

                    B0[J] = temp_val_1 * (1.0 / temp_val_2);
                    B1[J] = (B0[J] - B0[KINDEX]) / TAULS[J];


                # The very top of the atmosphere is isothermal
                B1[0] = 0;

                # This solves for the C values in the toon code
                for J in range(NEW_NLAYER):
                    if(0 > J-1):
                        KINDEX = 0;
                    else:
                        KINDEX = J-1;

                    temp_gamma_val[J]   = 1.0 / (y1[J] + y2[J]);

                    CP[J]  = (B0[KINDEX] + B1[J] * temp_gamma_val[J]) * 2.0 * PI * mu_1;
                    CPB[J] = CP[J] + B1[J] * TAULS[J] * 2.0 * PI * mu_1;

                    CM[J]  = (B0[KINDEX] - B1[J] * temp_gamma_val[J]) * 2.0 * PI * mu_1;
                    CMB[J] = CM[J] + B1[J] * TAULS[J] * 2.0 * PI * mu_1;

                J = 0;
                for L in range(1, 2*NEW_NLAYER-1, 2):
                    E[L]   = (CP[J+1] - CPB[J]) * e2[J+1] + (CMB[J] - CM[J+1]) * e4[J+1];
                    E[L+1] = e3[J] * (CP[J+1] - CPB[J]) + e1[J] * (CMB[J] - CM[J+1]);
                    J = J + 1;


                # HERE ARE THE TOP AND BOTTOM BOUNDARY CONDITIONS AS WELL AS THE
                # BEGINNING OF THE TRIDIAGONAL SOLUTION DDINITIONS. I ASSUME NO
                # DIFFUSE RADIATION IS INCIDENT AT THE TOP.

                E[0] = -CM[0];
                E[2*NEW_NLAYER-1]  = SFCS_HEMISPHERIC + RSFX * CMB[NEW_NLAYER-1] - CPB[NEW_NLAYER-1];
                DS[2*NEW_NLAYER-1] = E[2*NEW_NLAYER-1] / B[2*NEW_NLAYER-1];
                AS[2*NEW_NLAYER-1] = A[2*NEW_NLAYER-1] / B[2*NEW_NLAYER-1];


                #for(L=2; L<2*NEW_NLAYER+1; L++)
                for L in range(2, 2*NEW_NLAYER+1):
                    X[2*NEW_NLAYER-L]  = 1.0 / (B[2*NEW_NLAYER-L] - D[2*NEW_NLAYER-L] * AS[2*NEW_NLAYER-L+1]);
                    AS[2*NEW_NLAYER-L] = A[2*NEW_NLAYER-L] * X[2*NEW_NLAYER-L];
                    DS[2*NEW_NLAYER-L] = (E[2*NEW_NLAYER-L] - (D[2*NEW_NLAYER-L] * DS[2*NEW_NLAYER-L+1])) * X[2*NEW_NLAYER-L];


                Y[0] = DS[0];
                #for(L=1; L<2*NEW_NLAYER; L++)
                for L in range(1, 2*NEW_NLAYER):
                    Y[L] = DS[L] - AS[L] * Y[L-1];

                #********************************************
                #*    Source Function Technique Solution    *
                #********************************************
                #for(J=1; J<NEW_NLAYER+1; J++)
                for J in range(1, NEW_NLAYER+1):
                    SOURCE_Y1[J-1] = Y[2*J-2];
                    SOURCE_Y2[J-1] = Y[2*J-1];


                for J in range(NEW_NLAYER):
                    if(0 > J-1):
                        KINDEX = 0;
                    else:
                        KINDEX = J-1;

                    SOURCE_G[J] = (SOURCE_Y1[J] + SOURCE_Y2[J]) * (1.0/mu_1 - LAMBDAS[J]);
                    SOURCE_H[J] = (SOURCE_Y1[J] - SOURCE_Y2[J]) * GAMMA[J] * (1.0/mu_1 + LAMBDAS[J]);
                    SOURCE_J[J] = (SOURCE_Y1[J] + SOURCE_Y2[J]) * GAMMA[J] * (1.0/mu_1 + LAMBDAS[J]);
                    SOURCE_K[J] = (SOURCE_Y1[J] - SOURCE_Y2[J]) * (1.0/mu_1 - LAMBDAS[J]);

                    source_temp[J] = (1.0 / (y1[J] + y2[J])) - mu_1;

                    ALPHA_1[J]     = 2 * PI * (B0[J] + (B1[J] * source_temp[J]));
                    ALPHA_2[J]     = 2 * PI * (B1[J]);

                    SIGMA_1[J] = 2.0 * PI * (B0[J] - (B1[J] * source_temp[J]));
                    SIGMA_2[J] = 2.0 * PI * B1[J];


                INTENSITY_DOWN[0] = BB_TOP_OF_ATM * np.exp(-TAULS[0]) + \
                                  SOURCE_J[0]/(LAMBDAS[0] + 1.0) * (1.0 - np.exp(-TAULS[0] * (LAMBDAS[0]+1.0) )) + \
                                  SOURCE_K[0]/(LAMBDAS[0] - 1.0) * (np.exp(-TAULS[0]) - np.exp(-TAULS[0]*LAMBDAS[0])) + \
                                  SIGMA_1[0] * (1.0 - np.exp(-TAULS[0])) + \
                                  SIGMA_2[0] * (np.exp(-TAULS[0]) + TAULS[0] + 1.0);

                # Do the downward intensity first
                #for(J=1; J<NEW_NLAYER; J++)
                for J in range(1, NEW_NLAYER):
                    INTENSITY_DOWN[J] = INTENSITY_DOWN[J-1] * np.exp(-TAULS[J]) + \
                                    SOURCE_J[J]/(LAMBDAS[J] + 1.0) * (1.0 - np.exp(-TAULS[J] * (LAMBDAS[J]+1.0) )) + \
                                    SOURCE_K[J]/(LAMBDAS[J] - 1.0) * (np.exp(-TAULS[J]) - np.exp(-TAULS[J]*LAMBDAS[J])) + \
                                    SIGMA_1[J] * (1.0 - np.exp(-TAULS[J])) + \
                                    SIGMA_2[J] * (np.exp(-TAULS[J]) + TAULS[J] - 1.0);

                INTENSITY_UP[NEW_NLAYER-1] = 2.0 * BB_BOTTOM_OF_ATM * EMIS * PI;
                intensity = 0;

                # Calculate the upward intensity next
                #for(Z=1; Z<NEW_NLAYER; Z++)
                for Z in range(1, NEW_NLAYER):
                    J = NEW_NLAYER - Z - 1;

                    INTENSITY_UP[J] = INTENSITY_UP[J+1] * np.exp(-TAULS[J+1]) + \
                                      SOURCE_G[J] / (LAMBDAS[J]-1.0) * (np.exp(-TAULS[J+1])-np.exp(-TAULS[J+1] * (LAMBDAS[J]))) + \
                                      SOURCE_H[J] / (LAMBDAS[J]+1.0) * (1.0 - np.exp(-TAULS[J+1] * (LAMBDAS[J] + 1.0))) + \
                                      ALPHA_1[J] * (1.0 - np.exp(-TAULS[J+1])) + \
                                      ALPHA_2[J] * (1.0 - ((TAULS[J+1] + 1.0) * (np.exp(-TAULS[J+1]))))


                # Boundary conditions
                mu_1 = 0.577350;
                EMIS = 0.0;
                RSFX = 0.0;
                SFCS_QUADRATURE = RSFX * mu_0 * np.exp(-(TAUCS[NEW_NLAYER-1]) / mu_0) * PI * FLUX_SURFACE_QUADRATURE;

                # HERE WE FIND LAYER PROPERTIES FOLLOWING GENERAL SCHEME
                # OF MEADOR AND WEAVOR. THEN WE SET UP LAYER PROPERTIES
                # NEEDED FOR MATRIX.

                for J in range(NEW_NLAYER):
                    y1[J] = 0.86602540378 * (2.0 - (W0[J] * (1.0 + G0[J])));
                    y2[J] = 0.86602540378 * W0[J] * (1.0 - G0[J]);

                    y3[J] = 0.5 * (1.0 - 1.73205 * mu_0 * G0[J]);
                    y4[J] = 1.0 - (y3[J]);

                    LAMBDAS[J]    =  (y1[J] ** 2.0 - y2[J] ** 2.0) ** 0.5
                    GAMMA[J]  =  y2[J] / (y1[J] + LAMBDAS[J]);
                    temp_e_val[J]   =  np.exp(-LAMBDAS[J] * TAULS[J]);

                    e1[J]   =  1.0 + GAMMA[J] * temp_e_val[J];                         
                    e2[J]   =  1.0 - GAMMA[J] * temp_e_val[J];                    
                    e3[J]   =  GAMMA[J] + temp_e_val[J];                      
                    e4[J]   =  GAMMA[J] - temp_e_val[J];
                    

                J = 0;
                for L in range(1, 2*NEW_NLAYER-1, 2):
                    # HERE ARE THE EVEN MATRIX ELEMENTS
                    A[L]   =  e2[J+1] * e1[J]   - e3[J] * e4[J+1];
                    B[L]   =  e2[J+1] * e2[J]   - e4[J+1] * e4[J];
                    D[L]   =  e1[J+1] * e4[J+1] - e3[J+1] * e2[J+1];

                    # HERE ARE THE ODD MATRIX ELEMENTS EXCEPT FOR THE TOP. 
                    A[L+1] =  e2[J]   * e3[J]   - e1[J]   * e4[J];
                    B[L+1] =  e1[J+1] * e1[J]   - e3[J+1] * e3[J]; 
                    D[L+1] =  e3[J]   * e4[J+1] - e1[J]   * e2[J+1];
                    J = J + 1;


                # HERE ARE THE TOP AND BOTTOM BOUNDARY CONDITIONS AS WELL AS THE
                # BEGINNING OF THE TRIDIAGONAL SOLUTION DDINITIONS. I ASSUME
                # NO DIFFUSE RADIATION IS INCIDENT AT UPPER BOUNDARY.
                A[0] = 0.0;
                B[0] = e1[0];
                D[0] = -e2[0];

                A[2*NEW_NLAYER-1] = e1[NEW_NLAYER-1] - RSFX * e3[NEW_NLAYER-1];
                B[2*NEW_NLAYER-1] = e2[NEW_NLAYER-1] - RSFX * e4[NEW_NLAYER-1];
                D[2*NEW_NLAYER-1] = 0.0;



                for J in range(NEW_NLAYER):
                    if(0 > J-1):
                        KINDEX = 0;
                    else:
                        KINDEX = J-1;

                    temp_gamma_val[J]   = 1.0 / (y1[J] + y2[J]);

                    CP[J] =  W0[J] * PI * FLUX_SURFACE_QUADRATURE * \
                             np.exp(-(TAUCS[J]) / mu_0) * \
                             (((y1[J] - 1.0 / mu_0) * y3[J]) + (y4[J] * y2[J])) / \
                             (LAMBDAS[J] ** 2.0 - (1.0 / (mu_0 ** 2.0)));


                    CPB[J] = W0[J] * PI * FLUX_SURFACE_QUADRATURE * \
                             np.exp(-(TAUCS[J] + TAULS[J]) / mu_0) *   \
                             (((y1[J] - 1.0 / mu_0) * y3[J]) + (y4[J] * y2[J])) / \
                             (LAMBDAS[J] ** 2.0 - (1.0 / (mu_0 ** 2.0)));


                    CM[J] = W0[J] * PI * FLUX_SURFACE_QUADRATURE * \
                            np.exp(-(TAUCS[J]) / mu_0) *   \
                            (((y1[J] + 1.0 / mu_0) * y4[J]) + (y2[J] * y3[J])) / \
                            ((LAMBDAS[J] ** 2.0) - (1.0 / (mu_0 ** 2.0)));


                    CMB[J] = W0[J] * PI * FLUX_SURFACE_QUADRATURE * \
                             np.exp(-(TAUCS[J] + TAULS[J]) / mu_0) *   \
                             (((y1[J] + 1.0 / mu_0) * y4[J]) + (y2[J] * y3[J])) / \
                             ((LAMBDAS[J] ** 2.0) - (1.0 / (mu_0 ** 2.0)));


                J = 0;
                for L in range(1, 2*NEW_NLAYER-1, 2):
                    E[L]   = (CP[J+1] - CPB[J]) * e2[J+1] + (CMB[J] - CM[J+1]) * e4[J+1];
                    E[L+1] = e3[J] * (CP[J+1] - CPB[J]) + e1[J] * (CMB[J] - CM[J+1]);
                    J = J + 1;

                E[0] = -CM[0];
                E[2*NEW_NLAYER-1]  = SFCS_HEMISPHERIC + RSFX * CMB[NEW_NLAYER-1] - CPB[NEW_NLAYER-1];
                E[2*NEW_NLAYER-1]  = 0;

                DS[2*NEW_NLAYER-1] = E[2*NEW_NLAYER-1] / B[2*NEW_NLAYER-1];
                AS[2*NEW_NLAYER-1] = A[2*NEW_NLAYER-1] / B[2*NEW_NLAYER-1];

                for L in range(2, 2*NEW_NLAYER+1):
                    #print (2*NEW_NLAYER-L)
                    X[2*NEW_NLAYER-L]  = 1.0 / (B[2*NEW_NLAYER-L] - D[2*NEW_NLAYER-L] * AS[2*NEW_NLAYER-L+1]);
                    AS[2*NEW_NLAYER-L] = A[2*NEW_NLAYER-L] * X[2*NEW_NLAYER-L];
                    DS[2*NEW_NLAYER-L] = (E[2*NEW_NLAYER-L] - (D[2*NEW_NLAYER-L] * DS[2*NEW_NLAYER-L+1])) * X[2*NEW_NLAYER-L];

                Y[0] = DS[0];
                for L in range(1, 2*NEW_NLAYER):
                    Y[L] = DS[L] - AS[L] * Y[L-1];

                    
                for J in range(NEW_NLAYER):
                    TMI[J] = (1.0 / mu_1) * (Y[2*J]*(e1[J] + e3[J]) + \
                             (Y[2*J+1] * (e2[J] + e4[J])) + CPB[J] + CMB[J]);
                    TMI[J] = np.abs(TMI[J])
                    
                TMI = np.asarray(TMI)
                QUADRATURE_TWO_STREAM = np.asarray(QUADRATURE_TWO_STREAM)
                
                intensity = 0
                intensity_list = []
                for i in range(NEW_NLAYER):
                    def get_intensity(temperature, NU):
                        temp_val_1 = (2.0 * h_constant * (NU * NU * NU)) / (CLIGHT * CLIGHT);
                        temp_val_2 = np.exp(h_constant * NU / (bolz_constant * temperature)) - 1.0;
                        return temp_val_1 * (1.0 / temp_val_2);
                
                    intensity += get_intensity(TEMPS[i], NU) * np.exp(-TAUCS[i]) * TAULS[i]
                    intensity_list.append(intensity)
                
                THERMAL_FLUXES[W0_G0].append(intensity_list[-1])
                FLUXES[W0_G0].append(INTENSITY_UP[0] / (2 * PI) + TMI[0] / (4.0 * PI))
```
